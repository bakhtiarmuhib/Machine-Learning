```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    total_cost = 0
    for i in range(m):
        cost = ((x[i]*w+b)-y[i])**2
        total_cost += cost
    return total_cost/(2*m)
```

```python
cost = compute_cost(x_train,y_train,100,100)
print(cost)
```

```python
x = np.array([1,2,3,4])
y = np.array([1,2,3,4])
w = np.array([-2,-1.5,-1,-.5,0,1,1.5,2,2.5,3,3.5,4])
cost_for_w = np.zeros(len(w))
```

```python
for i in range(len(w)):
    cost_for_w[i]=compute_cost(x,y,w[i],0)
```

```python
style.use('ggplot')
plt.figure(figsize=(10,8))
plt.scatter(w,cost_for_w, marker='x',c='r')
plt.plot(w,cost_for_w,c='b')
plt.axhline(0)
plt.axvline(0)
plt.xlim(-3,5)
plt.ylim(-1,40)
plt.show()
```
