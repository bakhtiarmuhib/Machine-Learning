```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
print(f"x_tain is: {x_train}")
print(f"y_tain is: {y_train}")
```

```python
# m is number of training example
print(f"x_train.shape:{x_train.shape}")
m = x_train.shape[0]
print(f"Number of training example:{m}")
```

```python
style.use('ggplot')
plt.figure(figsize=(10,5))
plt.scatter(x_train,y_train, marker='x', c='r')
plt.title('house price')
plt.xlabel("size")
plt.ylabel('price')
plt.show()
```

```python
w = 100
b = 100
print(f"Weigth:{w}")
print(f"Bias:{b}")
```

```python
def compute_model_output(x,w,b):
    m = x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i]+b
    print(f_wb)
    return f_wb
```

```python
temp_f_wb = compute_model_output(x_train,w,b)

style.use('ggplot')
plt.figure(figsize=(10,5))
plt.scatter(x_train,y_train,marker='x',c='r', label="Actual price")
plt.plot(x_train,temp_f_wb,c='b',label="Our Prediction")
plt.title("Housing price")
plt.xlabel("Size in (1000 sqf)")
plt.ylabel("price in (1000 k)")
plt.legend()
plt.show()
```
